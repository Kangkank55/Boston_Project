```python
# Compute correlation matrix
correlations = boston.corr()

# Get correlation values with target column
target = 'median home price'
target_corr = correlations[target]

# Loop through all features and plot if correlation is strong
for feature in boston.columns:
    if feature != target:
        corr_value = target_corr[feature]
        if abs(corr_value) >= 0.5:  # You can adjust this threshold
            plt.figure(figsize=(6, 4))
            sns.regplot(x=boston[feature], y=boston[target], line_kws={"color": "red"})
            plt.title(f"{feature} vs {target} (corr={corr_value:.2f})")
            plt.xlabel(feature)
            plt.ylabel(target)
            plt.tight_layout()
            plt.show()
```

```python
boston
```

```python
#Independent and dependent features
X=boston.iloc[:,:-1]
y=boston.iloc[:,-1]
```

```python
X.head()
```

```python
y
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
# Standardize the dataset
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

 **Model** **Training:**

```python
# model training
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
# print the coefficients
print(lr.coef_)
```

```python
print(lr.intercept_)
```

```python
# on which parameters the model has been trained
lr.get_params()
```

```python
# prediction with test data
pred=lr.predict(X_test)
```

```python
pred
```

```python
# plot a scatter plot for the prediction
plt.scatter(y_test,pred)
```

```python
# calculate errors and plot
residuals=y_test-pred
residuals.plot.hist()
```

```python
residuals
```

```python
sns.displot(residuals,kind='kde')
```

```python
# scatter plot with respect to residual and predictions
plt.scatter(x=pred,y=residuals)
# uniformly distributed
```

```python
# perfomance matrix
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
print('MAE:',mean_absolute_error(y_test,pred))
print('MSE:',mean_squared_error(y_test,pred))
print('RMSE:',np.sqrt(mean_squared_error(y_test,pred)))
```

```python
# R square and adjusted R square
print('R square value:',r2_score(y_test,pred))
print('Adjusted R square value:',1-(1-r2_score(y_test,pred))*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1))
```
